```python
#120. If they enter a 1, it should run a subprogram that will  generate two random numbers between 5 and 20, and  ask the user to add them together. Work out the correct  answer and return both the user’s answer and the  correct answer.  If they entered 2 as their selection on the menu, it  should run a subprogram that will generate one number  between 25 and 50 and another number between 1 and  25 and ask them to work out num1 minus num2. This  way they will not have to worry about negative answers.  Return both the user’s answer and the correct answer.  Create another subprogram that will check if the user’s  answer matches the actual answer. If it does, display  “Correct”, otherwise display a message that will say  “Incorrect, the answer is” and display the real answer.  If they do not select a relevant option on the first menu  you should display a suitable message. 
import random
def addition():
    y = random.randint(5, 20)
    u = random.randint(5, 20)
    i = y, " + ", u, " = "
    print(i)
    user_answer = int(input("What do you think the answer is? "))
    check_answer = y + u
    return (user_answer, check_answer)

def subtraction():
    y = random.randint(5, 20)
    u = random.randint(5, 20)
    i = y, " - ", u, " = "
    print(i)
    user_answer = int(input("What do you think the answer is? "))
    check_answer = y - u
    return (user_answer, check_answer)

def check(user_answer, check_answer):
    if user_answer == check_answer:
        print("Correct!")
    elif user_answer != check_answer:
        print("Incorrect!")

def main():
    print("1. Addition")
    print("2. Subtraction")
    x = int(input("Enter a 1 or a 2. "))
    if x == 1:
        user_answer, check_answer = addition()
        check(user_answer, check_answer)
    if x == 2:
        user_answer, check_answer = subtraction()
        check(user_answer, check_answer)
main()

    
```

```python
#127. #127. Create a window that will ask the user to enter a  name in a text box. When they click on a button it  will add it to the end of the list that is displayed on  the screen. Create another button which will clear  the list. 

from tkinter import * 
def add_name():
    name = name_box.get()
    name_list.insert(END, name)
    name_box.delete(0, END)
    name_box.focus()
    
def clear_list():
    name_list.delete(0, END)
    name_box.focus()
    
window = Tk()
window.title("Names list")
window.geometry("400x200")

label1 = Label(text = "Enter a name: ")
label1.place(x=20, y=20, width = 100, height = 25)

name_box = Entry(text = 0)
name_box.place(x=120, y=20, width = 100, height = 25)
name_box.focus()

butt1 = Button(text = "Add to list", command = add_name)
butt1.place(x = 250, y=20, width = 100, height = 25)

name_list = Listbox()
name_list.place(x=120, y = 50, width = 100, height = 100 )

but = Button(text = "Clear list", command = clear_list)
but.place(x=250, y=50, width=100, height=25)
window.mainloop()
```

```python
#128. 1 kilometre = 0.6214 miles and 1  mile = 1.6093 kilometres. Using  these figures, make a program  that will allow the user to  convert between miles and  kilometres. 
from tkinter import *
def kilo():
    kilo = kilo_box.get()
    kilo = float(kilo)
    message = kilo * 0.6214
    butt1.delete(0, END)
    butt1.insert(END, message)
    butt1.insert(END, "miles")
    butt1.focus()
    
    
def miles():
    miles = miles_box.get()
    miles = float(miles)
    message1 = miles * 1.60934
    but.delete(END, message)
    but.insert(END, message1)
    but.insert(END, "km")
    but.focus()

window = Tk()
window.title("Interchange kilometers and miles. ")
window.geometry("400x200")

label1 = Label(text = "Enter a unit to be changed: ")
label1.place(x=20, y=20, width = 200, height = 25)

kilo_box = Entry(text = "")
kilo_box.place(x=30, y=50, width = 200, height = 25)
kilo_box["justify"] = "center"
kilo_box.focus()

miles_box = Entry(text = "")
miles_box.place(x=30, y=50, width = 200, height = 25)
miles_box["justify"] = "center"
miles_box.focus()

butt1 = Button(text = "Convert km to miles.   ", command = kilo)
butt1.place(x = 250, y=20, width = 200, height = 25)


but = Button(text = "Convert miles to km.  ", command = miles)
but.place(x=250, y=50, width=200, height=25)

window.mainloop()
```
